We import requests, so we can fetch data on database using API

```python
import requests
iso_codes={}
```

We associate each country form the excel database with it's alphacode

```python
for index, name in df['Country (according to CN Gov):'].items():
    if name in iso_codes:
        continue
    url = 'https://restcountries.eu/rest/v2/name/'+ name
    r=requests.get(url)
    if r.status_code==200:
        data=r.json()
        country_code=data[0]['alpha3Code']
        iso_codes[name]=country_code
    else:
        print('[ERROR] we still canot get the data, reason is',r.status_code,r.text)
        cleaned_name = name.split('(')[0].strip()
        print(name,cleaned_name)
```

Correction of the error encountered during the precedent step, with the typo mistake and the "(Not BRI)"

```python
for index, name in df['Country (according to CN Gov):'].items():
    if name in iso_codes:
        continue
    cleaned_name = name.split('(')[0].strip() 
    if cleaned_name == 'Thailiand':
        cleaned_name='Thailand'
    elif cleaned_name == 'The United Arab Emirates':
        cleaned_name='United Arab Emirates'
    
    url = 'https://restcountries.eu/rest/v2/name/'+ cleaned_name
    r=requests.get(url)
    if r.status_code==200:
        data=r.json()
        country_code=data[0]['alpha3Code']
        iso_codes[name]=country_code
    else:
        print('[ERROR] we still canot get the data')
        print(name,cleaned_name)
```

Creating the list of the data we need from Worldbank, and fetching them using the iso code and the data code + correcting "air freight"

```python
indic = {}
table = {'IMF WB\nCustoms and other import duties (% of tax revenue) 2016': 'GC.TAX.IMPT.ZS', 'WB IMF Non-performing loans to total gross loans ratio (%)': 'FB.AST.NPER.ZS', 'UIC Rail freight (mn ton x km travelled)': 'IS.RRS.GOOD.MT.K6', 'ICAO air freight (mn ton x km travelled)': 'IS.AIR.GOOD.MT.K1', 'WB Household final consumption expenditure, PPP bn USD': 'NE.CON.PRVT.PP.CD', 'WB UNESCO R&D Expenditure % of GDP': 'GB.XPD.RSDV.GD.ZS', 'WB UNESCO Researchers p. mn inhabitants': 'SP.POP.SCIE.RD.P6', 'IFDI Net Inflows (% of GDP)': 'BX.KLT.DINV.WD.GD.ZS', 'WHO Current Health Expenditure p.c., PPP USD': 'SH.XPD.CHEX.PP.CD', 'WB Research Expenditure % of GDP': 'GB.XPD.RSDV.GD.ZS', 'WIPO Trademark applications': 'IP.TMK.RSCT',
'WIPO Patent Applications': 'IP.PAT.RESD'}
for colonne, codeindic in table.items():
    if colonne not in indic:
        indic[colonne] = {}
    print(codeindic)            
    for index, isocode in df["isocode"].items():    
        url = 'http://api.worldbank.org/v2/country/' + isocode + '/indicator/'+ codeindic +'?mrv=1&format=json'
        r = requests.get(url)
        
        if r.status_code == 200:
            print(isocode)
            data = r.json()        
            try:
                indic[colonne][isocode] = data[1][0]['value']
                print(indic[colonne][isocode])
            except:
                indic[colonne][isocode] = "NA"
                print(indic[colonne][isocode])
        else:
            print('[ERROR] we  canot get the data')
```
